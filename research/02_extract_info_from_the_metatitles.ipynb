```python
import spacy; from spacy import displacy;
```

```python
nlp= spacy.load("en_core_web_trf")
```

```python
def render_line(line):
    doc= nlp(line)
    displacy.render(
        docs= doc,
        style= 'ent'
    )
```

```python
def extract_entities_from_keyword(keyword: str, num_results: int= 10):
    metatitle= generate_seo_metatitle(keyword, num_query_results=num_results)
    # print(metatitle)
    render_line(metatitle)
```

```python
extract_entities_from_keyword("hybrid cars", 50)
```

```python
extract_entities_from_keyword("mercedes benz")
```
