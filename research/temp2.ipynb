```python
import spacy; from spacy import displacy;
```

```python
txt= "Razer Kraken Headphones are the best headphones in the market."
```

```python
nlp= spacy.load("en_core_web_trf")
```

```python
doc= nlp(txt)

displacy.render(
    docs=doc,
    style='ent'
)
```

```python
def render_line(line):
    doc= nlp(line)
    displacy.render(
        docs= doc,
        style= 'ent'
    )
```
